```python
import numpy as np
```

```python
def gauss(a):
  n = a.shape[0]
  det = 1.0
  det_steps = []
  iterasi = 1

  for k in range(n):
    pivot = a[k, k]
    a[k, k:] /= pivot
    det *= pivot
    det = round(det, 2)
    a = np.round(a, 2)

    det_steps.append(round(pivot, 2))

    print(f"Iterasi ke-{iterasi}:")
    print(f"b{k + 1} = b{k + 1} / {pivot}")
    print(a)
    iterasi += 1

    for i in range(k + 1, n):
      factor = a[i, k]
      a[i, k:] -= factor * a[k, k:]
      a = np.round(a, 2)
      print(f"Iterasi ke-{iterasi}:")
      print(f"b{i + 1} = b{i + 1} - ({factor}) * b{k + 1}")
      print(a)
      iterasi += 1

  det_formula = " × ".join(map(str, det_steps)) + f" = {det}"
  return a, det_formula
```

```python
a = [[7, -5, -7], [-5, 9, 9], [4, -5, 3]]
a = np.array(a, float)
print("Matriks Awal:")
print(np.round(a, 2))
print()

A, DET = gauss(a)

if A is not None and DET is not None:
  print("\nMatriks Akhir:")
  print(A)
  print(f"\nDeterminan = {DET}")
```
